```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('KNNAlgorithmDataset.csv')
dataset.head()
```

```python
dataset.shape
dataset.info()
```

```python
dataset = pd.get_dummies(data=dataset,drop_first=True,columns=['diagnosis'])
```

```python
dataset.drop(columns=['id','Unnamed: 32'],axis=1,inplace=True)
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
corr = dataset.corr()
fig, ax = plt.subplots(figsize=(18, 16))
sns.heatmap(data=corr,cmap='coolwarm',vmax=1,vmin=-1,annot=True,ax=ax)
```

```python
dataset.drop(columns=['fractal_dimension_mean','texture_se','smoothness_se','symmetry_se'],axis=1,inplace=True)
```

```python
cols_to_scale = dataset.drop(columns=['diagnosis_M']).columns
```

```python
for col in cols_to_scale:
    dataset[col] = (dataset[col]-dataset[col].mean())/dataset[col].std()
```

```python
dataset['concavity_mean'].std()
```

```python
X = dataset.drop(['diagnosis_M'],axis=1)
Y = dataset['diagnosis_M']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=.2)
```

```python
# x_train.shape
# x_test.shape
# y_train.shape
# y_test.shape
```

```python
class KNN:
    def __init__(self,n_neighbours,n):
        self.n_neighbours = n_neighbours
        self.N = n
    def euclidian_distance(self,p1,p2):
        dist = 0
        for i in range(self.N):
            dist += (p2[i]-p1[i])**2
        return np.sqrt(dist)
    def get_neighbours(self,x_train,y_train,x_test):
        distances = []
        for i in range(len(x_train)):
            distances.append((y_train[i],self.euclidian_distance(x_train[i],x_test)))
        distances.sort(key=lambda x:x[1])
        neighbours = []
        for i in range(self.n_neighbours):
            neighbours.append(distances[i][0])
        return neighbours

    def fit_predict_for_one_row(self,x_train,y_train,x_test):
        dict = {}
        neighbours = self.get_neighbours(x_train,y_train,x_test)
        s = set()
        for i in y_train:
            s.add(i)
        for i in s:
            dict[i] = 0
        for i in range(self.n_neighbours):
            dict[neighbours[i]] += 1
        mx = -1
        label = y_train[0]
        for  i in dict.keys():
            if mx < dict[i]:
                mx = dict[i]
                label = i
        return label

    def fit_predict(self,x_train,y_train,x_test):
        x_train = x_train.values.tolist()
        y_train = y_train.values.tolist()
        x_test = x_test.values.tolist()
        predictions = []
        for row in x_test:
            predictions.append(self.fit_predict_for_one_row(x_train,y_train,row))
        return predictions
```

```python
model_knn = KNN(n_neighbours=3,n=x_train.shape[1])
```

```python
y_predicted = model_knn.fit_predict(x_train,y_train,x_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test,y_predicted))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn2 = KNeighborsClassifier()
model_knn2.fit(x_train,y_train)
y_predicted2 = model_knn2.predict(x_test)
```
